## 이진 탐색

### 비재귀적 함수

```python
def binary_search_iterative(arr, x):
    low = 0
    high = len(arr) - 1
    comparisons = 0  # 비교 횟수 초기화

    while low <= high:
        mid = (high + low) // 2
        comparisons += 1  # 비교 횟수 증가
        print(f"Comparison {comparisons}: arr[{mid}] = {arr[mid]}")  # 비교 단계 출력

        if arr[mid] == x:
            print(f" {mid} 번 인덱스에 있음. {comparisons} 회 비교함.")             
            return mid
        elif arr[mid] < x:
            low = mid + 1
        else:
            high = mid - 1

    print(f" 없음. {comparisons} 회 비교함.")
    return None

        

#테스트

arr = [6, 11, 12, 14, 16, 34, 42, 43, 44, 48, 49, 60, 63, 68, 76, 81, 88, 91, 95, 98]
key = 68
binary_search_iterative(arr, key)

arr = [2, 4, 3, 10, 10, 10, 20]
x = 10
binary_search_iterative(arr, x)
```

## 이진 탐색
### 재귀적 구현

```python
def binary_search_recursive(arr, low, high, x, comparisons=0):
    if low <= high:
        mid = (high + low) // 2
        comparisons += 1  # 비교 횟수 증가
        if arr[mid] == x:
            print(f" {mid} 번 인덱스에 있음. {comparisons} 회 비교함.")
            return mid, comparisons
        elif arr[mid] > x:
            return binary_search_recursive(arr, low, mid - 1, x, comparisons)
        else:
            return binary_search_recursive(arr, mid + 1, high, x, comparisons)
    else:
        print(f" 없음. {comparisons} 회 비교함.")
        return None, comparisons


#테스트
arr = list(range(100000000)) #1억 개의 리스트 : 0 ~ 99999999
key = len(arr) #100000000
binary_search_recursive(arr, 0, len(arr) - 1, key)
```

### 정렬되지 않은 큰 리스트를 퀵 정렬 후 랜덤한 수를 이진 탐색한다

```python
# 퀵정렬 구현
def quick_sort(arr):
    if len(arr) <= 1:
        return arr
    pivot = arr[len(arr) // 2]
    left = [x for x in arr if x < pivot]
    middle = [x for x in arr if x == pivot]
    right = [x for x in arr if x > pivot]
    return quick_sort(left) + middle + quick_sort(right)

# 이진탐색 구현 (재귀적)
def binary_search_recursive(arr, low, high, x, comparisons=0):
    if high >= low:
        mid = (high + low) // 2
        comparisons += 1  # 비교 횟수 증가
        print(f"Comparison {comparisons}: arr[{mid}] = {arr[mid]}")  # 비교 단계 출력

        if arr[mid] == x:
            print(f"Element found at index {mid} after {comparisons} comparisons.")
            return mid, comparisons
        elif arr[mid] > x:
            return binary_search_recursive(arr, low, mid - 1, x, comparisons)
        else:
            return binary_search_recursive(arr, mid + 1, high, x, comparisons)
    else:
        print(f"Element not found after {comparisons} comparisons.")
        return None, comparisons
```

```python
import random
# 정렬되지 않은 큰 리스트 생성
arr = [random.randint(1, 2000) for _ in range(100)]
x = random.choice(arr)
print("Original array:", arr)
print("key:", x)
print()
# 퀵정렬 수행
sorted_arr = quick_sort(arr)
print("Sorted array:", sorted_arr)
print()

# 이진탐색 수행
start = 0
end = len(sorted_arr) - 1
key = x
binary_search_recursive(sorted_arr, start, end, key)
```

### 이진 탐색을 응용하여 리스트에서 같은 수가 나타나는 구간 인덱스 얻기

```python
def find_first_last(arr, x):
    first = binary_search_recursive_first(arr, 0, len(arr)-1, x)
    last = binary_search_recursive_last(arr, 0, len(arr)-1, x)
    return first, last

def binary_search_recursive_first(arr, low, high, x):
    if high >= low:
        mid = (high + low) // 2
        if (mid == 0 or arr[mid-1] < x) and arr[mid] == x:
            return mid
        elif arr[mid] < x:
            return binary_search_recursive_first(arr, mid + 1, high, x)
        else:
            return binary_search_recursive_first(arr, low, mid - 1, x)
    else:
        return None

def binary_search_recursive_last(arr, low, high, x):
    if high >= low:
        mid = (high + low) // 2
        if (mid == len(arr)-1 or x < arr[mid+1]) and arr[mid] == x:
            return mid
        elif x < arr[mid]:
            return binary_search_recursive_last(arr, low, mid - 1, x)
        else:
            return binary_search_recursive_last(arr, mid + 1, high, x)
    else:
        return None

# 테스트
arr = [1, 4, 10, 10, 10, 10, 20]
x = 10
first, last = find_first_last(arr, x)
print(f"First occurrence of {x}: Index {first}")
print(f"Last occurrence of {x}: Index {last}")
```
