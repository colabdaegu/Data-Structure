# 퀵 정렬

```python
# 분할
def partition(S, start, end):
    pivot = S[start]  # 피벗
    i = start + 1     # 더 큰 수를 찾는 인덱스
    j = end           # 더 작은 수를 찾는 인덱스
    while i <= j:
        while i <= j and S[i] <= pivot: i += 1
        while i <= j and S[j] >= pivot: j -= 1
        if i < j:
            S[i], S[j] = S[j], S[i]  # swap
            i += 1
            j -= 1
    pivot_index = j
    S[start], S[j] = S[j], S[start]
    
    return pivot_index

# 재귀호출
def quicksort(S, start, end):
    if start < end:
        pivot_index = partition(S, start, end) #분할 함수 호출
        # 분할되는 부분 리스트 출력
        print(f'분할: {S[start:pivot_index]}, {S[pivot_index]}, {S[pivot_index+1:end+1]}')  
        quicksort(S, start, pivot_index - 1) # 왼쪽 리스트 정렬
        quicksort(S, pivot_index + 1, end)   # 오른쪽 리스트 정렬

S = [5,7,9,0,3,4]
S = [15, 10, 25, 8, 30, 5]
print("정렬 전:", S)
quicksort(S, 0, len(S) - 1)
print("정렬 후:", S)
```
