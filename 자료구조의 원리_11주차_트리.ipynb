### 트리 노드 클래스

```python
class TNode: 
    def __init__(self,val):
        self.val = val
        self.left = None
        self.right = None 
```

### 트리 순회 함수

```python
def preorder(n):
        if n is not None :
            print(n.val, end=' ')
            preorder(n.left)
            preorder(n.right)
                
def inorder(n):
        if n is not None :
            inorder(n.left)
            print(n.val, end=' ')
            inorder(n.right)
            
def postorder(n):
        if n is not None:
            postorder(n.left)
            postorder(n.right)
            print(n.val, end=' ')               
```

### 트리 생성 및 순회 함수 테스트

```python
root = TNode('A')
B = TNode('B')
C = TNode('C')
root.left = B
root.right = C

D= TNode('D')
E= TNode('E')
B.left = D
B.right = E

F= TNode('F')
G= TNode('G')
C.left = F
C.right = G

H= TNode('H')
I= TNode('I')
D.left = H
D.right = I

print('\n  전위 순회 : ', end=' ')
preorder(root)

root = TNode('77')
B = TNode('51')
C = TNode('97')
root.left = B
root.right = C

D= TNode('36')
E= TNode('65')
B.left = D
B.right = E

F= TNode('82')
G= TNode('99')
C.left = F
C.right = G

H= TNode('14')
I= TNode('39')
D.left = H
D.right = I

print('\n  전위 순회 : ', end=' ')
preorder(root)
print('\n  중위 순회 : ', end=' ')
inorder(root)
print('\n  후위 순회 : ', end=' ')
postorder(root)
```

### 최소 힙

```python
import heapq as hq  #최소 힙 라이브러리, hq로 사용

test = [8, 5, 3, 1, 3, 4, 9, 2, 5, 6, 7]
min_heap = [] #최소 힙 리스트

for i in test:
    hq.heappush(min_heap, i) # 최소 힙 생성

print(min_heap) # 최소 힙 출력
hq.heappush(min_heap, 0) #0을 삽입
print(min_heap) # 0 삽입 후 출력
print("최소 값: ", hq.heappop(min_heap)) #최소 힙에서 값 삭제
print(min_heap)
print("최소 값: ", hq.heappop(min_heap))  #최소 힙에서 값 삭제
print(min_heap)
```

```python
# 최대 힙 리스트
max_heap = []
test = [8, 5, 3, 1, 3, 4, 9, 2, 5, 6, 7]
# 최대 힙 구현: 값을 음수로 변환하여 최소 힙을 최대 힙처럼 사용
for i in test:
    hq.heappush(max_heap, -i)  # 값을 음수로 변환하여 삽입

# 최대 힙 출력 (원래 값을 복원하기 위해 다시 음수로 변환)
print([-i for i in max_heap])

# 100을 삽입 (음수로 변환하여 삽입)
hq.heappush(max_heap, -100)  
print([-i for i in max_heap])  # 최대 힙 출력

# 최대 값 출력 (음수로 저장되어 있으니 다시 원래 값으로 돌려서 출력)
print("최대 값: ", -hq.heappop(max_heap))  # 최대 힙에서 값 삭제
print([-i for i in max_heap])  # 현재 최대 힙 상태 출력

# 다시 최대 값 삭제
print("최대 값: ", -hq.heappop(max_heap))  # 최대 힙에서 값 삭제
print([-i for i in max_heap])  # 현재 최대 힙 상태 출력
```
